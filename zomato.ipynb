# Zomato Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataFrame = pd.read_csv('Zomato data.csv')
```

```python
dataFrame.head()
```

```python
def handleRate(value):
  value = str(value).split('/')
  value = value[0]
  return float(value)
```

```python
dataFrame['rate'] = dataFrame['rate'].apply(handleRate)
```

```python
dataFrame.info()
```

**Q1: TYPE OF RESTAURANT DO THE MAJORITY OF CUSTOMER ORDER FROM ** ****

```python
dataFrame.head()
```

```python
sns.countplot(x=dataFrame['listed_in(type)'])
plt.xlabel('RESTAURANT TYPE')
plt.ylabel('COUNT')
plt.title('TYPE OF RESTAURANT DO THE MAJORITY OF CUSTOMER ORDER FROM')
plt.show()
```

**Q2)Total Votes each restaurante has recived**

```python
grouped_data = dataFrame.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result, marker ="o")
plt.xlabel('RESTAURANT TYPE', size=10)
plt.ylabel('TOTAL VOTES',size=10)
```

**Q3) Ratings that majority of reataurant has recived**




```python
dataFrame.head()
```

```python
plt.hist(x=dataFrame['rate'])
plt.xlabel('Rates')
plt.ylabel('Count')
plt.title('Rating Distribution')
plt.show()
```

**Q4) Average spending of couples on each order**

```python
couples = dataFrame['approx_cost(for two people)']
sns.countplot(x=couples)
plt.xlabel('Cost')
plt.ylabel('Count')
plt.title('Average spending of couples on each order')
plt.show()
```
